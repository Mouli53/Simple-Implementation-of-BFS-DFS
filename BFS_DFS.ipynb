```python
def DFS(G,s):
    S = []
    out = []
    S.append(s)
    G[s][1] = True
    while len(S)!=0:
        v = S[-1] 
        out.append(S.pop())
        for w in G[v][0][::-1]:
            if G[w][1]==False:
                S.append(w)
                G[w][1] = True
    return out
```

```python
def BFS(G,s):
    Q = []
    out = []
    Q.append(s)
    G[s][1] = True
    while len(Q)!=0:
        v = Q[0]
        out.append(Q.pop(0))
        for w in G[v][0]:
            if G[w][1]==False:
                Q.append(w)
                G[w][1] = True
    return out
```
